```python
import pandas as pd
import numpy as np
from scipy.stats import t
```

```python
#given data from the question
sample_mean=3050
theoritical_mean=1000+5*600
standard_deviation=5*25
sample_size=25
alpha=0.05
```

```python
#calculating t_statistics
standard_error=standard_deviation/np.sqrt(sample_size)
t_statistics=(sample_mean-theoritical_mean)/standard_error
```

```python
print("standard_error :",standard_error)
print("t_statistics:",t_statistics)
```

```python
dof=sample_mean-1      #degree of freedom
dof
```

```python
#determining the critical value
critical_value=t.ppf(1-alpha,dof)
critical_value
```

```python
#Decision making
if t_statistics<critical_value:
    print("Accept H0:The operating cost for a franchise in a week is not changed and retained same as $1000+$5X for X=600")
else:
    print("Reject H0:The operating cost for a franchise in a week is changed as greater than $1000+$5X for X=600")
```
