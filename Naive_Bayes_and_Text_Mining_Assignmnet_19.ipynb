```python
import re
def clean_text(text):
  text = re.sub(r'[^a-zA-Z\s]', '', text)    #Remove special characters and punctuation
  text = text.lower() # Convert to lowercase
  text = text.strip()  #Remove leading and trailing whitespace
  return text
df['cleaned'] = df['Data'].apply(clean_text)
df
```

```python
#tokenization:
#Tokenize the cleaned text data and store the tokens in a new column

from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

def tokenize_text(text):
  return word_tokenize(text)

df['tokens'] = df['cleaned'].apply(tokenize_text)
df
```

```python
#Stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

df['tokens_no_stopwords'] = df['tokens'].apply(remove_stopwords)
df
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(word):
    """Map POS tag to first character used by WordNetLemmatizer"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"N": wordnet.NOUN,
                "V": wordnet.VERB,
                "A": wordnet.ADJ,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]

df['lemmatized_tokens'] = df['tokens_no_stopwords'].apply(lemmatize_tokens)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x=vectorizer.fit_transform(df["cleaned"])
y=df["Labels"]
```

```python
#Naive Bayes Model For Text Classification
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42,stratify=y)
```

```python
nb=MultinomialNB()
nb.fit(X_train,Y_train)
y_pred=nb.predict(X_test)
```

```python
print("Accuracy:",accuracy_score(Y_test, y_pred))
print("Precision:",precision_score(Y_test, y_pred, average="macro", zero_division=0))
print("Recall:",recall_score(Y_test, y_pred, average="macro", zero_division=0))
print("F1-score:",f1_score(Y_test, y_pred, average="macro", zero_division=0))
print("Classification Report:",classification_report(Y_test, y_pred,zero_division=0))
```

```python
from textblob import TextBlob

def get_sentiment(text):
  polarity = TextBlob(str(text)).polarity
  if polarity > 0.1:
    return "positive"
  elif polarity < -0.1:
    return "negative"
  else:
    return "neutral"
```

```python
df['Sentiment'] = df['Data'].apply(get_sentiment)
sentiment_counts = df['Sentiment'].value_counts()
print("Sentiment distribution:\n", sentiment_counts)
```

```python
# Sentiment by category
sentiment_by_category = df.groupby('Labels')['Sentiment'].value_counts().unstack(fill_value=0)
print("Sentiment by category:\n", sentiment_by_category)
```
