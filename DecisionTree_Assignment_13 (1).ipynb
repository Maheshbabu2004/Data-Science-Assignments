```python
#3. Feature Engineering
```

```python
#Converting bool to int
df['fbs'] = df['fbs'].astype(int)
df
```

```python
df.dtypes
```

```python
for column in df.columns:
  if df[column].dtype == 'object':
    df[column] = df[column].astype(str)
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_val=df.select_dtypes(include='object').columns
le=LabelEncoder()
for column in categorical_val:
  df[column]=le.fit_transform(df[column])
```

```python
df.rename(columns={'num':'target'},inplace=True)
df
```

```python
#correlation heatmap
import seaborn as sns
corrlaton_matrix=df.corr()
plt.figure(figsize=(15,8))
sns.heatmap(corrlaton_matrix,annot=True)
plt.title('CORRELATION HEATMAP')
plt.show()
```

```python
#Task-4.Decision Tree classification:
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
import warnings
warnings.simplefilter('ignore')
```

```python
x=df.drop('target',axis=1)
y=df['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#Task5. Hyperparameter Tuning:
```

```python
from sklearn.model_selection import GridSearchCV
params={'criterion':['gini','entropy'],'max_depth':[1,2,3,4,5,6,7],'splitter':['random','best']}
grid=GridSearchCV(model,params)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
model2=DecisionTreeClassifier(criterion='gini',max_depth=3,splitter='random')
model2.fit(x_train,y_train)
```

```python
y_pred1=model2.predict(x_test)
print(classification_report(y_test,y_pred1))
```
