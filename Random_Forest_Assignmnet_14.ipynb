```python
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
a=df.values
x=a[:,0:9]
y=a[:,9]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
model = RandomForestClassifier(n_estimators=100, max_features=3)
model.fit(x_train,y_train)
prediction = model.predict(x_test)
prediction
```

```python
print(classification_report(y_test,prediction))
```

```python
# Bagging Method
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
a=df.values
X=a[:,0:9]
Y=a[:,9]
kfold=KFold(n_splits=10)
model=BaggingClassifier(DecisionTreeClassifier(), n_estimators=100, random_state=7)
results=cross_val_score(model,X,Y,cv=kfold)
print(results.mean())
```

```python
#. Boosting Method
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
a=df.values
X=a[:,0:9]
Y=a[:,9]
kfold=KFold(n_splits=10)
model=AdaBoostClassifier(n_estimators=10, random_state=7)
results=cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
#Bagging Showed More Accuracy than Boosting
```

```python
#1.Explain Bagging and Boosting methods. How is it different from each other


#Bagging and Boosting - these combine the predictions of multiple models to improve accuracy, but they differ fundamentally in how they construct and combine these models.

#Bagging- Reduces variance and helps prevent overfitting

###Example: Random forest

#Boosting- Reduces bias and can potentially improve the accuracy further than bagging.

#Example:AdaBoost,Gradient Boosting Machines, XGBoost
```
