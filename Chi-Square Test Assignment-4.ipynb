```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
#table data
data = np.array([[50,70],
                 [80,100],
                 [60,90],
                 [30,50],
                 [20,50]])
df=pd.DataFrame(data,columns=["Smart Thermostat", "Smart Light"],
                index=["Very Statisfied", "Satisfied", "Neutral", "Unsatisfied", "Very satisfied"])
df
```

```python
#Performing the chi square test of independence
chi2_value,p,dof,expected=chi2_contingency(data)
print("chi_square value:",chi2_value)
print("p_value:",p)
print("Degree of freedom",dof)
print("Expected frequency:",expected)
```

```python
#Finding the critical value
from scipy.stats import chi2
alpha=0.05     #alpha = 0.05 given in question
critical_value=chi2.ppf(1-alpha,dof)
print("Critical Value:", critical_value)
```

```python
if chi2_value < critical_value:
    print("Accept H0")
    print("There is no significant association between the type of smart home device purchased(smart thermostats vs. smart light) and the customer satisfaction level")
else:
    print("Reject H0")
    print("There is significant association between the type of smart home device purchased(smart thermostats vs. smart light) and the customer satisfaction level")
```
