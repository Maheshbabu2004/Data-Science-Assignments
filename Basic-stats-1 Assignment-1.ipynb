```python
import pandas as pd
df = pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
df.dtypes
```

```python
#Descriptive Analytics for Numerical columns
```

```python
numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
descriptive_stats = pd.DataFrame({
    'Mean': df[numerical_cols].mean(),
    'Median': df[numerical_cols].median(),
    'Mode': df[numerical_cols].mode().iloc[0],
    'Standard Dev': df[numerical_cols].std(),
})
```

```python
descriptive_stats
```

```python
print("Descriptive Statistics:\n")
print(descriptive_stats.round())
```

```python
#Data Visualization
```

```python
#histogram
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example: df is your DataFrame
#numerical_cols = ['Total sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Values']

# plot histogram for each numerical column
for col in numerical_cols:
  plt.figure(figsize=(8, 6))
  sns.histplot(df[col], bins=20, kde=True)
  plt.title(f'Histogram of {col}')
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.tight_layout()
  plt.show()
```

```python
#Boxplots:
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Loop through numerical columns
for col in numerical_cols:
  plt.figure(figsize=(8,6))
  sns.boxplot(x=df[col])
  plt.title(f'Boxplot of {col}')
  plt.xlabel(col)
  plt.tight_layout()
  plt.show()

```

```python
#Bar Chart Analysis for Categorical Column:
import matplotlib.pyplot as plt
import seaborn as sns

# select only categorical columns
categorical_cols = df.select_dtypes(include=['object', 'category'])

# plot bar graphs for each categorical column
for col in categorical_cols.columns:
  plt.figure(figsize=(10, 6))
  sns.countplot(data=df, y=col, order=df[col].value_counts().index, hue=col, palette='viridis')
  plt.title(f' Frequency of {col}')
  plt.xlabel('count')
  plt.ylabel(col)
  plt.tight_layout()
  plt.show()
```

```python
# Standarization of Numerical variables

from sklearn.preprocessing import StandardScaler

# select numerical columns for standardization
numerical_cols_to_standardize = df.select_dtypes(include=['number']).columns

#Initialize the standardScaler
scaler = StandardScaler()

#Apply the scaler to the numerical columns
df[numerical_cols_to_standardize] = scaler.fit_transform(df[numerical_cols_to_standardize])

print("\nDataframe after Standardization:")
df
```

```python
# Conversion of categorical Data into Dummy Variables
import pandas as pd
# Select only categotical columns for one-hot encoding
categorical_cols_to_encode = df.select_dtypes(include=['object', 'category']).columns

# Apply one-hot encoding to categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_cols_to_encode)

print("\nDataframe after One-Hot Encoding:")
df_encoded
```
