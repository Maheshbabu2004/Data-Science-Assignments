```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
df.head()
```

```python
X=df.drop("type",axis=1)
Y=df["type"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=40)
```

```python
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
prediction=model.predict(x_test) # predicting on test data set
prediction
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```

```python
from sklearn.model_selection import GridSearchCV
n_neighbors=np.array(range(1,40))
param_grid=dict(n_neighbors=n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid)
grid.fit(X, Y)
print("Best score:",grid.best_score_)
print("Best parameter:",grid.best_params_)
```

```python
#VISUALIZING THE DECISION BOUNDARIES OF THE CLASSIFIER
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
k_scores=[]
for k in range(1, 41):
    knn=KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(knn, X, Y, cv=5)
    k_scores.append(scores.mean())
plt.plot(range(1, 41), k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
### What are the key hyperparameters in KNN?

#Number of neighbors (k): This determines how many nearest neighbors are considered when making a prediction.

#Distance metric: The method used to calculate the distance between data points. Common metrics include Euclidean, Manhattan etc...

#Weights for neighbors: This decides how the contribution of each neighbor is considered, usually either uniform or distance.

#Algorithm type: This refers to how neighbors are searched, including brute-force and tree-based approaches such as-> KD-tree or Ball-tree, which affect computational efficiency but not the model's outcome
```

```python
### What Distance Metrics Can be Used In KNN?

#KNN uses various distance metrics to measure proximity between point. The choice of metric affects model performence and is ofen based on the nature of the data.

#Euclidean Distance: Measures straight-line distance in n-dimensional space. It's the default in many libraries and suitable for continuous, well-scaled data.

#Manhattan Distance: Sums the absolute differences across dimensions; works well with grid-like data and is less sensitive to outliers.

#Cosine Distance: Measures angle between vectors, focusing more on direction than magnitude; often used in text analytics and document similarity tasks.

#Hamming Distance: Counts differing positions in categorical or binary data

#Mahalanobis Distance: Accounts for correlations between variables and is useful for multivariate distributions
```
